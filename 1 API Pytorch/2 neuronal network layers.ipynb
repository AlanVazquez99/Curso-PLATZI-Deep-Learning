```python
from IPython.display import display

import torch
from torch import nn

torch.set_printoptions(sci_mode=False)
torch.set_default_tensor_type('torch.FloatTensor')
device = torch.device('cuda' if torch.cuda.is_available else 'cpu')
```

```python
linear = nn.Linear(in_features=4096, out_features=10) # Input dim = 1
conv2d = nn.Conv2d(in_channels=3, out_channels=6, kernel_size=3, stride=1, padding=1) # Modifica los canales, no modifica las dimensiones de la imagen
relu = nn.ReLU(False) # False para no modificar el tensor original
```

```python
x = torch.randn(4096)
y = linear(x)

display(x.shape)
display(y.shape)

display(x)
display(y)
```

```python
x = torch.randn((2, 3, 5, 7))
y = conv2d(x)
z = relu(x)

display(x.shape)
display(z.shape)
display(y.shape)

display(x)
display(z)
display(y)
```

```python
print(f'\nLINEAR')
for name, parameter in linear.named_parameters():
    print(name)
    print(parameter.shape, '\n')

print(f'\nCONV')
for name, parameter in conv2d.named_parameters():
    print(name)
    print(parameter.shape, '\n')

print(f'\nRELU')
for name, parameter in relu.named_parameters():
    print(name)
    print(parameter.shape, '\n')
```
